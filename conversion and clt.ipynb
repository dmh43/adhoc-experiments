```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as spst
```

```python
num_simulations = 10000
num_users = 10_000
cvr = 0.05
num_days = 30
converted = []
for sim_num in range(num_simulations):
    status = np.zeros(num_users, dtype=bool)
    for i in range(num_days):
        status |= np.random.rand(num_users) < cvr
    converted.append(status.sum())
```

```python
plt.hist(converted, bins=30)
```

```python
spst.probplot(converted, plot=plt)
```

```python
spst.shapiro(converted)
```
